```python
## Exercise 7.1A: Getting present data counts

pD="/Users/melyang/Desktop/PythonBootcamp2017/resources/"
indfile=open(pD+"51.2.2M.ind",'r') ##substitute archaic.ind
myindices=[(line.split()[2],ind) for ind,line in enumerate(indfile)]
indfile.close()

genofile=open(pD+"51.2.2M.geno",'r') ##substitute archaic.geno
mycounts=[0 for i in range(len(myindices))]
total=0
for line in genofile:
    total+=1
    for indiv,myindex in myindices:
        if line[myindex] in "012": mycounts[myindex]+=1
genofile.close()

print myindices
print mycounts
print total
```

```python
%matplotlib inline
## Exercise 7.1A continued - plotting in bar plot
## To avoid the amount of time it takes, I run the barplot in this new cell.
import matplotlib.pyplot as plt
import collections

mydat={i:mycounts[ind] for i,ind in myindices}
sortedset=collections.Counter(mydat).most_common()

fig=plt.figure(figsize=(12,6))
ax=fig.add_subplot(1,1,1)

mylabels = [i for i,count in sortedset]
x=[ind for ind in range(len(sortedset))]
mycountvals=[count for i,count in sortedset]

ax.bar(x,mycountvals,align='center')
ax.set_xlim(-1,len(mylabels))
ax.set_xticks(x)
ax.set_xticklabels(mylabels,rotation=90,fontsize=10)

ax.set_xlabel("Individual")
ax.set_ylabel("Available data (# Base Pairs)")
ax.set_title("Amt. Data Present")

plt.tight_layout()
plt.show()

print mylabels[-1], "has the most missing data, with only", mycountvals[-1], "bases called!"
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import random

x = range(-5,5)
ys = [[i for i in x],
    [i+3 for i in x],
    [-i for i in x],
    [-i+1 for i in x]]
ylabels=["x","x+3","-x","-x+1"]
fig = plt.figure()
axs=[fig.add_subplot(2,2,i+1) for i in range(4)]
axs[0].plot(x, ys[0], label=ylabels[0])
axs[1].plot(x, ys[1], 'bx', label=ylabels[1])
axs[2].plot(x, ys[2], 'y+', label=ylabels[2])
axs[3].plot(x, ys[3], '--', label=ylabels[3])
for i in range(4): 
    axs[i].set_xlim((-4,4))
    axs[i].set_ylim((-4,4))
    axs[i].set_xlabel("x")
    axs[i].set_ylabel(ylabels[i])
axs[0].legend(loc='upper left')
axs[1].legend(loc='lower right')
axs[2].legend(loc='lower left')
axs[3].legend(loc='lower left')

plt.tight_layout()
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import sequence_tools as st

## To avoid memory issues, I actually read each individual chromosome file in individually.
## This way, I don't get any errors related to memory. 

counts=[]
order="AGCT"

## For each chromosome get % of each base pair
for mychr in range(1,17):
    if mychr<10: chrname="0%s" % mychr
    else: chrname=str(mychr)
    mydat=st.fastaparser3("../resources/chr%s.fsa" % chrname)
    for mykey in mydat:
        chrcounts=[0,0,0,0] ##AGCT
        for bp in mydat[mykey]:
            for ind,setbp in enumerate(order):
                if bp==setbp: chrcounts[ind]+=1
        total=float(len(mydat[mykey]))
        chrcounts=[i/total for i in chrcounts]
        counts.append(chrcounts)

## Reorder so that we get four lists with % of the base pair for all chromosomes
countbybase=[]
for i in range(len(order)): 
    mybasecounts=[myset[i] for myset in counts]
    countbybase.append(mybasecounts)
print countbybase

##Plot!
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
for ind,i in enumerate(countbybase): 
    ax.plot(range(1,17),i,label=order[ind])
ax.set_xlabel("Chromosome Number")
ax.set_ylabel("Frequency of A/G/C/T")
ax.legend(loc="center right")
plt.tight_layout()
plt.show()
```
